## Create data

```python
X = np.random.normal(5,1,10000)
```

```python
sns.distplot(X)
```

```python
s = 1000
B = 1000

estimates = []
means = []

for i in range(B):
    sample = np.random.choice(X, size=s)
    x_bar = np.mean(sample)
    means.append(x_bar)
    estimates.append(np.exp(x_bar))
```

```python
sns.distplot(means)
```

```python
sns.distplot(estimates)
```
